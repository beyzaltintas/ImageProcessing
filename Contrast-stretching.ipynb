```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def apply_contrast_stretching(image_path):
    # Read the image
    original_image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

    # Normalize pixel values (bring them to the range of 0-255)
    normalized_image = cv.normalize(original_image, None, 0, 255, cv.NORM_MINMAX)

    # Apply contrast stretching
    contrast_stretched_image = cv.equalizeHist(normalized_image)

    # Visualize the images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(original_image, cmap='gray')
    plt.title('Original Image')

    plt.subplot(1, 3, 2)
    plt.imshow(normalized_image, cmap='gray')
    plt.title('Normalized Image')

    plt.subplot(1, 3, 3)
    plt.imshow(contrast_stretched_image, cmap='gray')
    plt.title('After Contrast Stretching')

    plt.show()

# Specify the image path
image_path = r'C:\Users\Beyza\Desktop\IP\img.jpg'

# Call the contrast stretching function
apply_contrast_stretching(image_path)
```
