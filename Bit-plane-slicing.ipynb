```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

```

```python
def bit_plane_slice(image):
    # Convert the image to grayscale
    gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

    # Get the dimensions of the image
    rows, cols = gray_image.shape

    # Visualize each bit plane
    for i in range(8):  # There are 8 bit planes
        bit_plane = np.bitwise_and(gray_image, 2 ** i)
        bit_plane = bit_plane * 255 // 2**i
        plt.subplot(2, 4, i+1)
        plt.imshow(bit_plane, cmap='gray')
        plt.title(f'Bit {i}')
        plt.xticks([]), plt.yticks([])

    plt.show()

# Read the image
image_path = r'C:\Users\Beyza\Desktop\IP\image.jpg'
image = cv.imread(image_path)

# Apply bit plane slicing
bit_plane_slice(image)
```
