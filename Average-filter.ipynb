```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv.imread(r'C:\Users\Beyza\Desktop\IP\img.jpg')

# Convert the image to grayscale
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Apply the average filter
kernel_size = 5
averaged_image = cv.blur(gray_image, (kernel_size, kernel_size))

# Visualize the images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(gray_image, cv.COLOR_GRAY2RGB))
plt.title('Original Grayscale Image')

plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(averaged_image, cv.COLOR_GRAY2RGB))
plt.title(f'Averaged Filter Applied Image (Kernel Size: {kernel_size})')

plt.show()
```
