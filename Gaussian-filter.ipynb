```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_gaussian_filter(image_path, kernel_size):
    # Read the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply the Gaussian filter
    gaussian_filtered_image = cv2.GaussianBlur(gray_image, (kernel_size, kernel_size), 0)

    # Visualize the images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB))
    plt.title('Original Grayscale Image')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(gaussian_filtered_image, cv2.COLOR_GRAY2RGB))
    plt.title(f'Gaussian Filter Applied Image (Kernel Size: {kernel_size})')

    plt.show()

# Specify the image path and Gaussian filter kernel size
image_path = r'C:\Users\Beyza\Desktop\IP\img.jpg'
kernel_size = 5  # You can change the kernel size as needed

# Call the Gaussian filter application function
apply_gaussian_filter(image_path, kernel_size)
```
