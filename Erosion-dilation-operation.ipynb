```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an example grayscale image
image = cv2.imread(r'C:\Users\Beyza\Desktop\IP\img.jpg', cv2.IMREAD_GRAYSCALE)

# Specify the operation to be performed (erosion or dilation)
operation = 'dilation'  # You can set it to 'erosion' or 'dilation'

# Define the size of the operation kernel
kernel_size = 5

# Apply the chosen operation
if operation == 'erosion':
    result_image = cv2.erode(image, np.ones((kernel_size, kernel_size), np.uint8), iterations=1)
elif operation == 'dilation':
    result_image = cv2.dilate(image, np.ones((kernel_size, kernel_size), np.uint8), iterations=1)
else:
    raise ValueError("Please specify a valid operation: 'erosion' or 'dilation'.")

# Visualize the images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(result_image, cmap='gray')
plt.title(f'{operation.capitalize()} Processed Image')

plt.show()
```
