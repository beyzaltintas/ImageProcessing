```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(image_path, gamma):
    # Read the image
    original_image = cv.imread(image_path)

    # Normalize and apply gamma correction
    gamma_corrected = np.power(original_image / 255.0, gamma) * 255.0
    gamma_corrected = np.uint8(gamma_corrected)

    # Visualize the images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(original_image, cv.COLOR_BGR2RGB))
    plt.title('Original Image')

    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(gamma_corrected, cv.COLOR_BGR2RGB))
    plt.title(f'Gamma Correction (γ={gamma})')

    plt.show()

# Specify the image path and gamma value
image_path = r'C:\Users\Beyza\Desktop\IP\img.jpg'
gamma_value = 1.5

# Call the gamma correction function
gamma_correction(image_path, gamma_value)
```
